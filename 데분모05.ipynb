```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./연도별_매출액_현황.csv')

data_total = df['전체']
data_general = df['일반통상']

plt.style.use('ggplot')

#1. figure()메소드를 사용하여 그림틀 객체를 만든다.
fig = plt.figure(figsize = (10,10))

#2. axe 객체 :: 한번에 여러 개의 그래프를 그려야 할 때 axe 객체를 사용한다.
            #:: 분할된 화면마다 axe 객체를 하나씩 배정함 .
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2) #여기서 fig는 위에서 그림틀 변수를 의미한다.
                             #(2,1,2) 2*1 행렬에서 2번째를 차지한다.

#3 각 axe객체별 그래프 생성 
ax1.plot(data_total, marker = 'o', markersize = 10, label = '전체') #plot에서 라벨을 지정할 수도 있다.
ax2.plot(data_general, marker = 'o', markersize = 10, markerfacecolor = 'green', \
        color = 'black', linewidth = 2, label = '일반통상')

ax1.legend(loc = 'best')
ax2.legend(loc = 'best')

ax1.set_ylim(0,600000)
ax2.set_ylim(5000,40000)

ax1.set_xticklabels(data_total.index, rotation = 60)
ax2.set_xticklabels(data_general.index, rotation = 60)
#그래프 한개일 때는 plt.xticks(rotation = 'vertical')
#한번에 그래프 여러개 할때는 위의 방식을 따른다. 

plt.show()


#그래프를 한번에 여러개
#1.fig = plt.figure(figsize = (10,10))
#2.ax1 = fig.add_subplot(2,1,1)
  #ax2 = fig.add_subplot(2,1,2)
#3.ax1.plot(시리즈, marker = 'o', markersize = , markerfacecolor = '', color = '', linewidth  = ,label = '')
#  ax2.plot(~)
#4.ax1.legend(loc = 'best')
#  ax2.legend(loc = 'best')
#5.ax1.set_ylim( , )
#  ax2.set_ylim( , )
#6.ax1.set_xtickslabel(시리즈.index, rotation = 60)
#  ax2.set_xtickslabel(시리즈.index, rotation = 60)
#7 plt.show()
```

## 4분할 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./연도별_매출액_현황.csv')

plt.style.use('ggplot')

#그래프 틀 
fig = plt.figure(figsize = (20,10))

#객체 생성 및 객체별 위치 결정 
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

#각 객체별 그래프 생성 

#ax1은 모든 설정을 지정 해봄. x축은 어떤것이 올지 y축은 어떤것이 올지도 지정함. 
ax1.plot(df.index, df['전체'], marker = 'o', markersize = 5, markerfacecolor = 'green', \
        color = 'black', linewidth = 2 , label = '전체')
ax1.legend(loc = 'best')

#x,y축 지정은 의무가 아님. 시리즈를 넣으면 알아서 index가 x축 값이 y축 됨
#ax2는 범례를 따로 지정해봤다.
ax2.plot(df['일반통상'], marker = 'o', markersize = 5, linewidth = 2)
ax2.legend(labels = ['일반통상'], loc = 'best')

#ax3,4는 기본
ax3.plot(df['국제특급'], marker = 'o', markersize = 5, linewidth = 2)
ax4.plot(df['소포'], marker = 'o', markersize = 5, linewidth = 2)

#title 정하기, 그래프 한번에 여러개 할때는 set_title함수를 사용 
ax1.set_title('전체')
ax2.set_title('일반통상')
ax3.set_title('국제특급')
ax4.set_title('소포')

#x축 값 각도 설정 , xlabel과는 다르다. 말그대로 축 설정임 .
ax1.set_xticklabels(df.index, rotation = 60)
ax2.set_xticklabels(df.index, rotation = 60)
ax3.set_xticklabels(df.index, rotation = 60)
ax4.set_xticklabels(df.index, rotation = 60)
```

```python
plt.style.use('ggplot')

fig = plt.figure(figsize=(16, 10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(df.index, df['전체'], marker='o', markersize=5, linewidth=2)
ax2.plot(df.index, df['일반통상'], marker='o', markersize=5, linewidth=2)
ax3.plot(df.index, df['국제특급'], marker='o', markersize=5, linewidth=2)
ax4.plot(df.index, df['소포'], marker='o', markersize=5, linewidth=2)

ax1.set_title('전체')
ax2.set_title('일반통상')
ax3.set_title('국제특급')
ax4.set_title('소포')

ax1.set_xticklabels(df.index, rotation=90)
ax2.set_xticklabels(df.index, rotation=90)
ax3.set_xticklabels(df.index, rotation=90)
ax4.set_xticklabels(df.index, rotation=90)

plt.show()
```

## 그래프 중첩

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./연도별_매출액_현황.csv')

#스타일 지정 
plt.style.use('ggplot')

#그래프 틀 지정
fig = plt.figure(figsize = (8,5))

#객체 지정 : 한 그래프 평면에 여러개의 그래프를 그리는 것이니, 객체는 한개만 필요.
ax = fig.add_subplot(1,1,1)

#각 그래프 생성

ax.plot(df.index, df['전체'], marker = 'o', markersize = 10, linewidth = 2, label = '전체')
ax.plot(df.index, df['소포'], marker = 'o', markersize = 10, linewidth = 2, label = '소포')
ax.plot(df.index, df['국제특급'], marker = 'o', markersize = 10, linewidth = 2, label = '국제특급')
ax.legend(loc = 'best')

ax.set_title('항목별 매출 현황', size = 20 )
ax.set_xlabel('연도', size = 10) #ax.set_xticklabels(df.index, rotation = )와 헷갈리지 말자.
ax.set_ylabel('백만원', size= 10)

ax.set_ylim(0,600000)

ax.set_xticklabels(df.index ,rotation = 60)

plt.show()

```

```python
#3가지 항목에 대한 면적 그래프를 그려보자

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./연도별_매출액_현황.csv')
df3 = df[['일반통상', '특수통상', '소포']]

#pandas 안에 있는 면적 그래프 사용하는 것임. df.plot()
#stacked = False : 중첩하지 않는다.
df3.plot(kind = 'area', stacked = False, alpha = 0.2, figsize= (10,5))

#title, xlabel, ylabel, label설정 

plt.title('항목별 매출 현황', size = 20)
plt.xlabel('연도', size = 10) #중첩이긴 하지만, 객체를 생성해서 하는게 아니기 때문에 xlabel로 진행한다.
plt.ylabel('백만원', size = 10)
plt.legend(loc = 'best') #plot 할때 label은 없었지만, 위의 형식으로 하면 저절로 label 설정됨.

plt.show()
```

# 막대그래프
- 막대 그래프는 데이터 값의 크기에 비례하여 높이를 갖는 직사각형 막대로 표현
- 세로형 그래프의 경우 선 그래프와 정보 전달 측면에서 큰 차이는 없음 
- 가로형 그래프의 경우 각 변수 사이 값의 크기 차이를 설명하는데 적합
    - 보통 값이 가장 큰 변수부터 내림차순 정렬하여 그래프를 그림 

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("./연도별_매출액_현황.csv", index_col = "연도")
#'연도'열을 인덱스 이름으로 지정해야한다.

df_3 = df[['일반통상', '특수통상', '소포']]
df_3.plot(kind = 'bar', width = 1.0, color = ['orange', 'green', 'red'],\
         figsize = (10,5))

plt.title("항목별 매출 현황", size = 20)
plt.xlabel("연도",size = 10)
plt.ylabel("백만원", size = 10)
plt.legend(loc = 'best')
plt.show()
```

# 세로형 막대 그래프

```python
#df.sum() 각 열의 합계를 더한다음 시리즈로 반환함.
df_sum = df.sum(axis = 0) #axis = 0 은 디폴트임 원래

#series.sort_values(ascending = True) 또는 df.sort_values(by = '열', ascending = True)
#기존 변수를 대체하고 싶다면 읽기와 쓰기부분에 동시에 배치 
df_sum = df_sum.sort_values(ascending = True)
print(df_sum)
```

```python
#수평막대그래프는 xy 바뀌는것 조심 
df_sum.plot(kind = 'barh', width = 0.5, color = 'blue', figsize = (8,4))

plt.title("항목별 매출 현황", size = 20)
plt.xlabel("백만원",size = 10)
plt.ylabel("항목", size = 10)
plt.show()

#각 항목별 매출 차이를 비교하기에 좋음 
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
iris.head()
```

- 구간의 개수(bins)에 따라 그래프 모양이 달라짐 

```python
plt.style.use('classic')

#당연 series.plot 가능함 .
#series.plot(kind = 'hist', bins = 10, color = , figsize = ())
iris['sepal_length'].plot(kind = 'hist', bins = 10, color = 'blue', figsize = (8,4))

plt.title("histogram")
plt.xlabel("sepal_length")
plt.show()

```
