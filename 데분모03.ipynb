## 판다스란?(전처리 도구)
- 데이터를 수집하고 정리하는데 최적화된 라이브러리
- 오픈소스로 무료라는 장점을 가짐
- 가장 쉬운 언어인 파이썬을 기반으로 함
- 데이터를 다루는 업무의 80% 이상을 판다스로 처리함
- 데이터 과학자에게 가장 기본적이면서 아주 중요한 도구 
- 판다스에서 제공하는 자료구조로는 **시리즈**, **데이터프레임**을 제공한다.
- 시리즈 : 1차원 배열, 데이터프레임 : 2차원(행,열)배열

## 시리즈(Series)
- 데이터가 순차적으로 나열된 1차원 배열 
- 인덱스(key)와 데이터 값(value)가 일대일 대응으로 이루어진 데이터
- 파이썬 딕셔너리와 비슷함.
- index(=key) - value 일대일 매칭 형식
- 딕셔너리는 key와 value이고 index가 없음. 그 차이
- 인덱스를 잘 활용하면 값의 탐색, 정렬, 선택, 결합을 쉽게할 수 있음.
- 정수형 인덱스가 기본으로 사용되며, 직접 이름(key)를 지정할 수도 있음.



- pd.Series(딕셔너리)
- pd.Series(리스트)
- pd.Series(리스트, index = 리스트)
- series.index, values
- series[[2,5]], series[['이름','나이']], series[2:5]

```python
#시리즈 생성 방법 1 : 딕셔너리를 시리즈로 변환 :: pd.Series(딕셔너리) :: 정수형인덱스(디폴트)=인덱스이름(딕셔너리 key) -  value 구조
import pandas as pd
dict_data = dict(zip(['a','b','c'],[1,2,3]))
series = pd.Series(dict_data)
print(type(series))
print(series)
print(series.index)
```

```python
#시리즈 생성 방법 2: 리스트를 시리즈로 변환
list_data = ['2021-04-26', '철수', '남', 100, True ]
series = pd.Series(list_data)

print(series)
print("\n")
print(series.index)
print("\n")
print(series.values)
```

```python
#시리즈 생성 방법3 : index 리스트 추가
lis_data = ['2021-04-26', '철수', '남', 100, True]
index_names = ['날짜', '이름', '성별', '점수', '전학여부']

series = pd.Series(list_data, index = index_names)
print(series)
print("\n")
print(series.index)
#정수형 인덱스는 디폴트로 무조건 있음
#인덱스이름은 옵션, 보통 인덱스이름 = 인덱스(index)로 칭해진다. 
#인덱스 이름을 따로 부과하면 인덱스 이름이 인덱스이고 출력은 인덱스 이름으로 출력됨
#따로부과하지 않으면 정수형인덱스로 출력됨
```

```python
#시리즈 원소 선택
lis_data = ['2021-04-26', '철수', '남', 100, True]
index_names = ['날짜', '이름', '성별', '점수', '전학여부']

series = pd.Series(list_data, index = index_names)
print(series)
print("\n")
print(series[0],series['이름']) #정수형 인덱스 또는 인덱스 이름을 시리즈 원소 선택 가능하다.
print("\n")
print(series[[0,2]]) #리스트와 다르게 시리즈에서는 [[ , ]]로 중복선택 가능함. 리스트는 (,)로 중복선택 불가
print("\n")
print(series[['날짜', '전학여부']])
print("\n")
print(series[2:5]) #정수형인덱스, 인덱스 이름으로 슬라이싱 가능함.
print("\n")
print(series['성별': '전학여부'])
```

## 데이터 프레임
- 2차원 배열로 행과 열로 이루어짐
- 여러개의 시리즈들이 모여서 데이터 프레임을 이룬다.


- pd.DataFrame({col : [원소]...})
- pd.DataFrame([[],[],[]...], index = [,], columns = [,])
- df.index = []
- df.columns = []
- df.rename(index = {before: after, .... })
- df.loc[], df.loc[[]]
- df.iloc[], df.iloc[[]]
- df[], df[[]]

```python
#데이터 프레임 만들기 1 : 딕셔너리로 만들기 
import pandas as pd
dic_data = {'col1' : [0,1,2],
            'col2' : [3,4,5],
            'col3' : [6,7,8],
            'col4' : [9,10,11]}
#{열이름 : [해당 열에 들어갈 원소] ,...} 구조

df = pd.DataFrame(dic_data) #pd.DataFrame(dic_data)
print(type(df))
df

```

```python
#행/열 이름 설정
#index : 행이름 
#columns : 열이름

#데이터프레임 만들기 2: 리스트를 활용
df = pd.DataFrame(
    [['남', 30, '경기고'],
     ['여', 25, '서울고']],
    index = ['기범','혜리'],
    columns = ['gender','age','school'])

#리스트로 이차원 배열 만들듯이 value를 채운다.
#index = [ , ] , columns = [ ] 형식으로 인덱스이름과 열의 이름을 정한다.

print(df)
print("\n")
print(df.index)
print("\n")
print(df.values)
```

```python
#행/열 이름 재설정 
#방법1 . 리스트를 활용 
print(df)
print('\n')
df.index = ['이름1', '이름2']
df.columns = ['성별', '연령', '학교']
print(df)
```

```python
#방법2. 딕셔너리를 활용한다.
print(df)
df.rename(columns = {'성별' : 'col1', '연령' : 'col2', '학교' : 'col3'})
df.rename(index = {'이름1' : 'index1', '이름2' : 'index2'})
print(df)
```

```python
#특정 행 선택방법
exam_data = {
    '국어' : [100,80,30],
    '수학' : [90,80,50],
    '영어' : [100,40,90],
    '음악' : [80,100,40]
}

#딕셔너리로 데이터프레임을 만들면서 index이름을 지정할 수 있다.
df = pd.DataFrame(exam_data, index = ['기범', '혜리', '동엽'])
print(df)
print("\n")
```

```python
#loc : 행 선택 :: 인덱스 이름으로 선택 
location = df.loc[['기범', '혜리']] #()아니고 , [[ , ]] (,)로 개별선택할 때는 반드시 [[,]] 
print(location)
print("\n")
ilocation = df.iloc[[1,2]]
print(ilocation) 
#시리즈는 행밖에 없기 때문에 series[]이런 구조이지만
#데이터프레임은 행과 열이기 때문에 df.loc or df.iloc[] 구조!
```

```python
#iloc고급활용
exam_data = {
    '국어' : [100,80,30,90],
    '수학' : [90,80,50,30],
    '영어' : [100,40,90,80],
    '음악' : [80,100,40,50]
}

df = pd.DataFrame(exam_data , index = ['기범', '혜리', '동엽', '나래'])
print(df)
print("\n")
df1 = df.iloc[1:3]
print(df1)
print("\n")
df2 = df.iloc[0:3:2] # 시작 끝 단계, 끝은 포함하지 않음.
print(df2)
print("\n")
df3 = df.iloc[::-1]
print(df3)
```

```python
#열 선택 
df = pd.DataFrame(
    {
    '국어' : [100,80,30],
    '수학' : [90,80,50],
    '영어' : [100,40,90],
    '음악' : [80,100,40], },
    index = ['기범','혜리','동엽']
)

print(df)
print("\n")

col1 = df['국어']  #행 선택은 loc 또는 iloc을 사용하지만 행선택은 그냥 df[] 이렇게 하면 된다.
                  #행 한개 선택 = 시리즈와 같음 
print(col1)
print("\n")

df1 = df[['국어','음악']] #행과 마찬가지로 ,로 중복선택하고 싶으면 반드시 [[ ]] 구조 
print(df1)
```

```python
#원소 선택

exam_data = {
    '국어' : [100,80,30,90],
    '수학' : [90,80,50,30],
    '영어' : [100,40,90,80],
    '음악' : [80,100,40,50]
}

df = pd.DataFrame(exam_data , index = ['기범', '혜리', '동엽', '나래'])
print(df)
print("\n")

elem1 = df.loc['기범', '국어'] # loc 또는 iloc [] 이중괄호가 아니라면 행, 열
print(f"기범의 국어성적은 {elem1}")

#혜리의 영어성적 iloc활용해서 
elem2 = df.iloc[1,2]
print("혜리의 영어성적은 {0}".format(elem2))
```

## 데이터프레임 조작
- 주어진 데이터프레임에 새로운 데이터를 추가하거나 삭제할 수 있음  (행, 열)
- 데이터프레임 내 특정 값이 잘못될 경우 이를 변경할 수 있음.
- 데이터 프레임의 행과 열의 위치를 바꿔서 조작할 수 있음
- 데이터 프레임 내 행 데이터를 특정 기준에 따라 정렬할 수 있음. 

- 행 추가
    - df.iloc[num] = [, , , ..] or df.loc[인덱스 이름]
    - df.loc['num'] = ~ 위에것과 비교 
    - 이미 존재하고 있는 행이름으로 해도 대체되는 것이 아니라 추가되는 것임.

- 열 추가
    - df[열이름] = 70
    - df[열이름] = [ , , , , ]
    - 이미 존재하고 있는 열이름으로 해도 대체되는 것이 아니라 추가되는 것임.
    
- 행, 열 삭제
    - df.drop('이름', axis = 0(1), inplace = True)
    - df.drop(['이름', , ] , axis = 0(1), inplace = True)
    
- 행/열 바꾸기
    - df.T or df.transpose()
    
- 정렬
    - df.sort_index(ascending = True)
    - df.sort_values(by = '열이름' , ascending = )
    


```python
#행 추가 
import pandas as pd
exam_data = {
    '이름' : ['기범', '혜리', '동엽'],
    '국어' : [100,80,30],
    '수학' : [90,80,50],
    '영어' : [100,40,90],
    '음악' : [80,100,40]
}

df = pd.DataFrame(exam_data)
print(df)
print("\n")

#loc을 활용한 행 추가
df.loc[4] = ['동현', 80,90,30,40] 
print(df)
print("\n")

df.loc[5] = df.loc[1] #앞에 있던 행을 중복 추가 가능함 
print(df)

#왜 loc인데 인덱스 이름을 하지 않았나?
#데이터 프레임에서 인덱스 이름은 선택사항.
#df.loc[4] 와 df.iloc[4]는 다르다.
#현재 df에서 df.loc[4]를 하면 정수형 인덱스 3에 4라는 인덱스 이름을 가진 새로운 행을 추가시키는 것 
#따라서 df.iloc[4]를 하면 정수형 인덱스가 차례로 나오지 않기 때문에 에러 발생함.

```

```python
#열 추가

import pandas as pd

exam_data = {
    '국어' : [100,80,30,90],
    '수학' : [90,80,50,30],
    '영어' : [100,40,90,80],
    '음악' : [80,100,40,50]
}

df = pd.DataFrame(exam_data , index = ['기범', '혜리', '동엽', '나래'])

print(df)
print("\n")

df['미술'] = 70 #미술 이라는 열은 없으니 미술 열을 새롭게 만들고 모든 원소를 70으로 채운다
print(df)
print("\n")

df['체육'] = [20,30,40,50]
print(df)
```

```python
#행삭제
exam_data = {
    '국어' : [100,80,30],
    '수학' : [90,80,50],
    '영어' : [100,40,90],
    '음악' : [80,100,40]
}

df = pd.DataFrame(exam_data , index = ['기범', '혜리', '동엽'])
print(df)
print("\n")

df2 = df[:] #열전체 선택 = 전체복사, ndarray와 다르게 df2를 변경하면 df도 변경되는 것 아님.
df2.drop('기범', inplace = True) #df.drop(행이름, inplace = True) , 행 index안됨 행이름으로 해야하고 inplace = True 인자를 넣어야함
print(df2)
print("\n")
df3 = df[:]
df3.drop(["기범", "동엽"], axis = 0 , inplace = True) #여러개의 행(열)을 (,)로 선택시 []리스트 형식으로, default는 axis = 0 (행)
print(df3)
```

```python
#열 삭제

exam_data = {
    '국어' : [100,80,30],
    '수학' : [90,80,50],
    '영어' : [100,40,90],
    '음악' : [80,100,40]
}

df = pd.DataFrame(exam_data , index = ['기범', '혜리', '동엽'])
print(df)
print("\n")

df2 = df[:]
df2.drop('국어', axis = 1 , inplace = True) #열 삭제는 반드시 인자로 axis = 1 있어야한다.
print(df2)
print("\n")

df3 = df[:]
df3.drop(['국어','음악'], axis = 1, inplace = True)
print(df3)
print("\n")
```

```python
#원소값 변경

import pandas as pd

exam_data = {
    '국어' : [100,80,30,90],
    '수학' : [90,80,50,30],
    '영어' : [100,40,90,80],
    '음악' : [80,100,40,50]
}

df = pd.DataFrame(exam_data , index = ['기범', '혜리', '동엽', '나래'])

print(df)
print("\n")

df.loc['혜리', '영어'] = 80 
df.iloc[0,3] = 30
print(df)

#데이터 프레임에서 원소에 접근하려면 loc또는 iloc [] 괄호 한개 구조로 한다.
#행, 열 추가는 중복이 가능하기에 이미 존재했던 걸 해도 중복되어 추가되지만
#원소 선택은 있는 자리를 지정하면 해당 원소가 대체된다.

```

```python
#행/열 위치바꾸기
#ndarray에서 1,2차원은 array.T 또는 array.transpose()
#ndarray에서 3차원 이상은 array.transpose(axis 정보)

#데이터 프레임에서는 ndarray 와 같이 df.T 또는 df.transpose()

import pandas as pd

exam_data = {
    '국어' : [100,80,30,90],
    '수학' : [90,80,50,30],
    '영어' : [100,40,90,80],
    '음악' : [80,100,40,50]
}

df = pd.DataFrame(exam_data , index = ['기범', '혜리', '동엽', '나래'])

print(df)
print("\n")

df = df.T
print(df)
print("\n")

df = df.transpose()
print(df)
print("\n")
```

```python
#데이터 프레임 정렬

dic_data = {
    'col1': [0,1,2],
    'col2': [5,4,3],
    'col3': [6,7,8],
    'col4': [9,11,10]
}

df = pd.DataFrame(dic_data, index = ['row1', 'row2', 'row3'])
print(df)
print('\n')

idf = df.sort_index(ascending = False) #df.sort_index(ascending = ) 인덱스이름을 기준으로 오름내림차순 정렬 
print(idf)
print("\n")

idf = df.sort_values(by = 'col4', ascending=True) #df.sort_values(by='열이름', ascending = ), 특정열을 기준으로 정렬
print(idf)
```
