### 특징
* 하나하나 칸들을 '셀'이라고 부른다
* 셀이 초록색이면 편집모드
* 셀이 파란색이면 명령모드
* 편집모드 상테에서 esc를 누르면 명령모드로 변환된다.
* 명령모드에서 enter을 누르면 편집모드로 전환된다.
* In[ num] 해당 셀의 실행순서를 의미한다.
* vscode에서 코드 실행은 위에서 아래로 순차적으로 진행된다.
* jupyter notebook에서는 내가 실행한 순서로 진행된다. 따라서 알 수 없는 오류가 있는 경우 ln[num]의 순서를 확인해보자.
* 모듈 , 자료형 등의 메소드 리스트를 확인하고 싶다면 모듈(자료형). 하고 tab을 누르면 사용가능한 메소드가 뜬다.
### 단축키
###### <shift + enter>
* 현재 셀을 실행하고 아래 셀이 없으면 새로운 셀을 만든다
* 아래 셀이 있다면, 현재 셀을 실행하고 아래셀을 클릭한 상태(명령모드)
###### <alt + enter>
* 아래 셀이 없다면 shift + enter와 같은 기능을 한다
* 아래 셀이 있다면 현재 셀을 실행하고 새로운 셀을 만든다.
###### <cntrl + enter>
* 현재 셀만 실행한다.
###### <d,d>
* 현재셀이 명령모드 상태에서 d를 두번 연속으로 누르면 해당 셀이 삭제된다.
###### < a >
* 셀 명령모드에서 a 하면 위에 빈셀 추가
###### < b >
* 아래 셀 추가
###### < i,i > 
* 인터럽트
###### < y > , < m >
* 셀 명령모드에서 실행하면 y : 코드로 전환 m : 마크다운으로 전환
###### < view - toggle line numbers>
* 셀별로 라인 넘버를 표시할 수 있다. (오류발생시 찾기 쉬움)
###### < kernel - restart & run all >
* 모든 셀을 위에서 아래로 순차적으로 실행(vscode처럼)
###### < file - download as  >
* html 파일로 저장하면 주피터노트북이 없는 환경에서도 실행결과를 동일하게 확인가능함.  
단, 직접 실행하고 수정하는 것은 불가능함
* 주피터 노트북을 저장하면 디렉토리에 .ipynb(interactive python notebook)이 생성되는데 이 파일을 넘기면  
주피터 노트북이 있는 사용자는 수정, 하나하나 실행이 모두 가능함 (file - open을 통해 열 수 있다)
